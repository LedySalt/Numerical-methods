## Initialization

```python
from matplotlib import pyplot as plt
import numpy as np

F = lambda x, y: 2*x*y/(x**2 + y**2)
x_0, y_0 = 1, 0.1
dx = 1e-2
n = 900
```

# Euler`s method

```python
def euler(F, x_0, y_0, dx = 1e-2, n = 1000):
  x, y = x_0, y_0
  data = [(x_0, y_0)]
  for i in range(n):
    y += dx*F(x, y)
    x += dx
    data.append((x, y))
  return data
```

# Plot

Euler`s plot

```python
points = np.array(euler(F, x_0, y_0, dx,n))
plt.plot(points[:, 0], points[:, 1], color='r', label=f'euler')
```

Exact plot

```python
from sympy import *
x = symbols('x')
y = Function('y')
F_= Eq(y(x).diff(x), 2*x*y(x)/(x**2 + y(x)**2))
equation = dsolve(F_, ics={y(10): 0.1})
plot(equation.rhs, (x, 1, 100))
```
